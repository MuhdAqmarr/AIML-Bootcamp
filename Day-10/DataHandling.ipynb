# Data Handling Using Pandas

```python
import pandas as pd
```

```python
df=pd.read_excel('students.xlsx')
print('Student Records as Follows')
df
```

```python
display(df)
```

```python
df.describe()
```

```python
avg_marks = df['Marks'].mean()
print(f'\nAverage marks: {avg_marks:.2f}')
```

```python
top_students=df[df['Marks']>80]
print('\nStudents with Marks > 80: ')
top_students
```

```python
rejected_students=df[df['Marks']<=80]
rejected_students
```

```python
top_students.to_excel('top_students.xlsx', index=False)
print('Filtered data saved to "top_students.xlsx"')
```

```python
rejected_students.to_excel('rejected_students.xlsx', index=False)
print('Filtered data saved to "rejected_students.xlsx"')
```

```python
import matplotlib.pyplot as plt
# Display Bar Chart
plt.figure(figsize=(10,5))
plt.bar(df['Name'],df['Marks'])

plt.title('Students Marks Data')
plt.xlabel('Students Name')
plt.ylabel('Marks')
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
# Display linechart for top students
plt.figure(figsize=(10,5))
plt.plot(top_students['Name'], top_students['Marks'], marker='o', linestyle='-', linewidth=2)

plt.title('Students scoring above 80 marks (LineChart)')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.grid(True, linestyle='--', alpha=0.6)
```

## Data from https://www.kaggle.com/datasets/jockeroika/life-style-data

```python
df2=pd.read_csv("Final_data.csv")
display(df2)
```

```python
df2.describe()
```

```python
selected_data = df2[['Age', 'Gender', 'Max_BPM']]
selected_data
```

```python
selected_data.describe()
```

**Filter data of persons whose weight greater than 60kg. We want only Age, Gender, Weight, protein_per_kg**

```python
select_data=df2[df2['Weight (kg)']>=60]
final_data=select_data[['Age', 'Gender', 'Weight (kg)','protein_per_kg']]
final_data
```

*Or we can write in a single line*

```python
final_data = df2.loc[df2['Weight (kg)'] >= 60, ['Age', 'Gender', 'Weight (kg)', 'protein_per_kg']]
final_data
```

```python
# Export data to excel
final_data.to_csv('overweight.csv', index=False)
print('Filtered data saved to "overweight.csv"')
```
